```python
import pandas as pd
import numpy as np
import re

PATH = "C:/Users/shata/OneDrive/Desktop/Buildinga job market intelligence platform/"

df1 = pd.read_csv(PATH + "data/raw/DataAnalyst.csv")
df2 = pd.read_csv(PATH + "data/raw/ds_salaries.csv")
df3 = pd.read_csv(PATH + "data/raw/naukri_com-job_sample.csv")
```

```python
# ══════════════════════════════════════════════════════════
# CLEAN DF1 — DataAnalyst.csv
# Only 1 missing value in Company Name → safe to drop
# ══════════════════════════════════════════════════════════

df1 = df1.drop_duplicates()
df1 = df1.dropna(subset=['Company Name'])          # drop that 1 missing row

# Parse salary: "$45K-$90K (Glassdoor est.)" → two numbers
def parse_salary(text):
    if pd.isna(text) or str(text).strip() == '-1':
        return np.nan, np.nan
    nums = re.findall(r'\d+', str(text))
    if len(nums) >= 2:
        return int(nums[0]) * 1000, int(nums[1]) * 1000
    return np.nan, np.nan

df1[['salary_min', 'salary_max']] = df1['Salary Estimate'].apply(
    lambda x: pd.Series(parse_salary(x))
)
df1['salary_avg'] = (df1['salary_min'] + df1['salary_max']) / 2

# Clean company name — remove rating appended after \n (e.g. "Google\n4.5")
df1['Company Name'] = df1['Company Name'].str.split('\n').str[0].str.strip()

df1_clean = df1[['Job Title', 'Company Name', 'Location',
                  'salary_min', 'salary_max', 'salary_avg',
                  'Job Description', 'Size', 'Industry', 'Sector']].copy()

df1_clean.columns = ['job_title', 'company', 'location',
                     'salary_min', 'salary_max', 'salary_avg',
                     'description', 'company_size', 'industry', 'sector']

df1_clean['source']   = 'DataAnalyst'
df1_clean['currency'] = 'USD'

print(f"✅ df1 cleaned: {df1_clean.shape}")
```

```python
# ══════════════════════════════════════════════════════════
# CLEAN DF2 — ds_salaries.csv
# No missing values at all — just rename & map codes
# ══════════════════════════════════════════════════════════

df2 = df2.drop_duplicates()

# Make experience level human-readable
exp_map = {'EN': 'Entry', 'MI': 'Mid', 'SE': 'Senior', 'EX': 'Executive'}
df2['experience_level'] = df2['experience_level'].map(exp_map)

# Make employment type human-readable
emp_map = {'FT': 'Full-Time', 'PT': 'Part-Time', 'CT': 'Contract', 'FL': 'Freelance'}
df2['employment_type'] = df2['employment_type'].map(emp_map)

# Map remote ratio numbers to labels
remote_map = {0: 'On-site', 50: 'Hybrid', 100: 'Remote'}
df2['work_mode'] = df2['remote_ratio'].map(remote_map)

df2_clean = df2[['job_title', 'experience_level', 'employment_type',
                  'salary_in_usd', 'company_size',
                  'employee_residence', 'company_location',
                  'remote_ratio', 'work_mode']].copy()

df2_clean.rename(columns={'salary_in_usd': 'salary_avg'}, inplace=True)
df2_clean['salary_min'] = (df2_clean['salary_avg'] * 0.85).astype(int)
df2_clean['salary_max'] = (df2_clean['salary_avg'] * 1.15).astype(int)
df2_clean['source']     = 'DS_Salaries'
df2_clean['currency']   = 'USD'

print(f"✅ df2 cleaned: {df2_clean.shape}")
```

```python
# ══════════════════════════════════════════════════════════
# CLEAN DF3 — Naukri
# Strategy per column:
#   company (4 missing)        → drop those rows
#   experience (4 missing)     → drop those rows
#   jobdescription (4 missing) → fill with "Not provided"
#   joblocation_address (501)  → fill with "Unknown"
#   skills (528 missing)       → fill with "Not listed"
#   payrate (97 missing)       → keep as NaN (we won't use it for ML)
#   education (1996 missing)   → fill with "Not specified"
#   numberofpositions (17536)  → drop column (too empty to be useful)
#   site_name (18013 missing)  → drop column (too empty to be useful)
#   postdate (23 missing)      → drop those rows
# ══════════════════════════════════════════════════════════

df3 = df3.drop_duplicates()

# Drop columns that are mostly empty (>50% missing)
df3 = df3.drop(columns=['numberofpositions', 'site_name'])

# Drop rows where critical fields are missing
df3 = df3.dropna(subset=['company', 'experience', 'postdate'])

# Fill non-critical missing values
df3['jobdescription']      = df3['jobdescription'].fillna('Not provided')
df3['joblocation_address'] = df3['joblocation_address'].fillna('Unknown')
df3['skills']              = df3['skills'].fillna('Not listed')
df3['education']           = df3['education'].fillna('Not specified')
# payrate stays NaN — we won't force-fill salary data

# Rename columns to clean standard names
df3_clean = df3.rename(columns={
    'jobtitle'            : 'job_title',
    'company'             : 'company',
    'joblocation_address' : 'location',
    'skills'              : 'skills',
    'experience'          : 'experience_required',
    'jobdescription'      : 'description',
    'industry'            : 'industry',
    'education'           : 'education',
    'payrate'             : 'salary_text',
    'postdate'            : 'post_date'
})

df3_clean['source']     = 'Naukri'
df3_clean['currency']   = 'INR'
df3_clean['salary_avg'] = np.nan   # salary is text range, not numeric

print(f"✅ df3 cleaned: {df3_clean.shape}")
```

```python

# ══════════════════════════════════════════════════════════
# SAVE all 3 to data/processed/
# ══════════════════════════════════════════════════════════

df1_clean.to_csv(PATH + "data/processed/analyst_clean.csv",  index=False)
df2_clean.to_csv(PATH + "data/processed/salaries_clean.csv", index=False)
df3_clean.to_csv(PATH + "data/processed/naukri_clean.csv",   index=False)

print("\n✅ All 3 files saved to data/processed/")
print(f"   analyst_clean  : {df1_clean.shape}")
print(f"   salaries_clean : {df2_clean.shape}")
print(f"   naukri_clean   : {df3_clean.shape}")
```
